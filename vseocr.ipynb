# Run vseocr

```
from google.colab import files
from vseocr import save_subtitles_to_file

# Function to handle file upload and process it
def handle_uploaded_file(file_content):
    # Save the uploaded file to a temporary location
    temp_file_path = 'uploaded_video.mp4'
    with open(temp_file_path, 'wb') as f:
        f.write(file_content)

    # Call your processing function
    save_subtitles_to_file(
        video_path=temp_file_path,
        file_path='example.srt',
        lang='en',
        sim_threshold=80,
        conf_threshold=50,
        use_fullframe=True,
        brightness_threshold=210,
        similar_image_threshold=1000,
        frames_to_skip=1
    )

    return 'File processed successfully'

# Prompt user to upload a file
uploaded = files.upload()

# Handle each uploaded file
for filename, file_content in uploaded.items():
    result = handle_uploaded_file(file_content)
    print(result)
```
